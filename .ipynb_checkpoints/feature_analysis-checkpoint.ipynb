### Overlapping Data

```python
# Extract data for each level
ndre_level1 = df[df['severity_level'] == 1]['ndre']
ndre_level2 = df[df['severity_level'] == 2]['ndre']
ndre_level3 = df[df['severity_level'] == 3]['ndre']
ndre_level4 = df[df['severity_level'] == 4]['ndre']

lci_level1 = df[df['severity_level'] == 1]['lci']
lci_level2 = df[df['severity_level'] == 2]['lci']
lci_level3 = df[df['severity_level'] == 3]['lci']
lci_level4 = df[df['severity_level'] == 4]['lci']
```

```python
# Count mean and median for each level - NDRE
median_level1_ndre = np.median(ndre_level1)
median_level2_ndre = np.median(ndre_level2)
median_level3_ndre = np.median(ndre_level3)
median_level4_ndre = np.median(ndre_level4)
print("Median Level 1 - NDRE :", median_level1_ndre)
print("Median Level 2 - NDRE :", median_level2_ndre)
print("Median Level 3 - NDRE :", median_level3_ndre)
print("Median Level 4 - NDRE :", median_level4_ndre)
```

```python
# Count mean and median for each level - LCI
median_level1_lci = np.median(lci_level1)
median_level2_lci = np.median(lci_level2)
median_level3_lci = np.median(lci_level3)
median_level4_lci = np.median(lci_level4)
print("Median Level 1 - LCI :", median_level1_lci)
print("Median Level 2 - LCI :", median_level2_lci)
print("Median Level 3 - LCI :", median_level3_lci)
print("Median Level 4 - LCI :", median_level4_lci)
```

Batasan nilai utk NDRE dan LCI (median):
1. Level 1 --> NDRE >= 0.19368204 dan LCI >= 0.28912251
2. Level 2 --> 0.17711057 < NDRE < 0.19368204 dan 0.25949356 < LCI < 0.28912251
3. Level 3 --> NDRE <= 0.17711057 dan LCI <= 0.25949356

```python
# Function to find overlapping data
def find_overlapping_data(val_ndre, val_lci):
    if (val_ndre >= median_level1_ndre) & (val_lci >= median_level1_lci):
        return 1
    elif (val_ndre <= median_level3_ndre) & (val_lci <= median_level3_lci):
        return 3
    else:
        return 2
```

```python
# Find overlapping data
list_overlapping_level = []
for i, row in df.iterrows():
    overlapping_level = find_overlapping_data(row['ndre'], row['lci'])
    list_overlapping_level.append(overlapping_level)
# Append to dataframe
df['overlapping_level'] = list_overlapping_level
```

```python
# Show data
df.head()
```

```python
# Find overlapping indexes data
idx_more1 = []
idx1 = []
idx_non = []
for i, row in df.iterrows():
    deviation = abs(row['severity_level'] - row['overlapping_level'])
    if (deviation > 1):
        idx_more1.append(i)
    elif (deviation == 1):
        idx1.append(i)
    else:
        idx_non.append(i)
```

```python
# Show indexes data
print("# of overlapping data > 1 Level :", len(idx_more1))
print("# of overlapping data = 1 Level :", len(idx1))
print("# of non-overlapping data :", len(idx_non))
```

#### Selisih Level > 1

```python
# Show data with level deviation > 1
df_more1 = df.loc[idx_more1]
df_more1
```

Total ada 9 data yang overlapping dengan selisih level lebih dari 1. Data-data dg selisih level > 1 dilabeli ulang dengan label pada kolom overlapping data

```python
# Replace old severity level with overlapping level
df_copy.loc[idx_more1, 'severity_level'] = df_more1.loc[idx_more1, 'overlapping_level']
df_copy.loc[idx_more1]
```
